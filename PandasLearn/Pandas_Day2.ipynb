```python
# Group by and get the count of employees in each department
df = pd.read_csv("../Data/emp.csv")
```

```python
df_grouped = df.groupby('deptno')['eno'].count().reset_index()
# rename the eno column to count
df_grouped.rename(columns={'eno':"countOfEmp"},inplace=True)
```

```python
# Get max,min,sum of the salary column per department
# sql : select sum(sal),min(sal),max(sal) ,count(sal) from emp group by deptno
df.groupby('deptno')['salary'].agg(['max','min','sum','count'])
```

```python
# Group by multiple column

df_grouped = df.groupby(['salary','deptno'])['eno'].count().reset_index()
```

```python
# Where clause equivalanet
df = pd.read_csv("../Data/emp.csv")
filter = df['deptno'] == 10
```

```python
df[filter]
```

```python
# to apply multiple deptno
filter_10_20 = (df['deptno']==10) | (df['deptno']==20)
```

```python
df[filter_10_20]
```

```python
# to apply more than one condition . e.g. deptno = 10 and salary = 1000
filter_deptno_sal = (df['deptno']==10) & (df['salary']==1000)
```

```python
df[filter_deptno_sal]
```

```python
# Having clause equivalent
df_grouped = df.groupby('deptno')['eno'].count().reset_index()
filter_having = df_grouped['eno'].apply(lambda x:x>8)
```

```python
df_grouped[filter_having]
```

```python
# Merge ( SQL join equivalent)
df_emp = pd.read_csv("../Data/emp.csv")
df_sal = pd.read_csv("../Data/sal.csv")

```

```python
#inner join
# SQL : select * from emp as e inner join sal as s on e.eno=s.eno
df_emp.merge(df_sal,on='eno',how='inner')
```

```python
df_emp
```

```python
df_emp
```

```python
#left join
# SQL : select * from emp as e left join sal as s on e.eno=s.eno
df_emp.merge(df_sal,on='eno',how='left')
```

```python
#right join
# SQL : select * from emp as e right join sal as s on e.eno=s.eno
df_emp.merge(df_sal,on='eno',how='right')
```

```python
#full outer join
# SQL : select * from emp as e full outer join sal as s on e.eno=s.eno
df_emp.merge(df_sal,on='eno',how='outer')
```

```python
# How to join 2 dataframews when joining keys have differet names

#inner join
# SQL : select * from emp as e inner join sal as s on e.eno=s.eno
df_emp.merge(df_sal,left_on='empno',right_on='eno',how='left')
```
